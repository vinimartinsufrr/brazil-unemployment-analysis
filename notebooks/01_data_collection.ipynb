```python
import requests
import pandas as pd
import os
```

```python
url = (
    "https://apisidra.ibge.gov.br/values/"
    "t/6381/"
    "n1/1/"
    "v/4099/"
    "p/all/"
    "d/v4099%202"
)
```

```python
response = requests.get(url)
response.raise_for_status()

data = response.json()
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df = df.iloc[1:].reset_index(drop=True)
```

```python
df = df[["D1N", "D3N", "V"]]

df.columns = [
    "territory",
    "period",
    "unemployment_rate"
]
```

```python
df["period"] = df["period"].astype(str)
df["unemployment_rate"] = df["unemployment_rate"].astype(float)
```

```python
import os

output_dir = "../data/raw"
os.makedirs(output_dir, exist_ok=True)

output_path = os.path.join(output_dir, "unemployment_rate_brazil.csv")
df.to_csv(output_path, index=False)

output_path

```
